Import `matplotlib` and `pandas` as `pd`

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Here we can see `minimum salaries` bar 

```python
df = pd.read_csv("job_data_cleaned.csv")
df_corrected = df[(df["min_salary"] > 360) | (df["max_salary"] > 360)]

plt.bar(df["title"], df["min_salary"], color="green")
plt.title("Min Salaries")
plt.xticks(rotation=90, fontsize=10)
plt.show()
min_sal = df_corrected["min_salary"].min()
print(f"Min Salary: {min_sal} UAH")
```

Lets take a `mean` value of `min salaries` 

```python
mean_for_min = int(df_corrected["min_salary"].mean())
print(f"Min Salary mean: {mean_for_min} UAH")
```

Here is `max salaries` bar

```python
plt.bar(df["title"], df["max_salary"], color="green")
plt.title("Max Salaries")
plt.xticks(rotation=90, fontsize=10)
plt.show()
max_sal = df_corrected["max_salary"].max()
print(f"Max Salary: {max_sal} UAH")
```

Mean for `max_salary`

```python
mean_for_max = int(df_corrected["max_salary"].mean())
print(f"Max Salary mean: {mean_for_max} UAH")
```

Lets compare `min` and `max` salaries 

```python
min_salary = df["min_salary"].min()
max_salary = df["max_salary"].max()

plt.plot(df.index, df["min_salary"], color="blue", marker="o", label="Min Salary")

# Plot the max_salary line
plt.plot(df.index, df["max_salary"], color="green", marker="o", label="Max Salary")

plt.title("Min, Max salaries comparison")
plt.xlabel("Index")
plt.ylabel("Salary")
plt.grid(True)

plt.legend()

# Display min and max values on the plot
plt.annotate(f'Min Salary: {min_salary}', xy=(0, min_salary), xytext=(0, min_salary + 5000),
             arrowprops=dict(facecolor='blue', shrink=0.05))
plt.annotate(f'Max Salary: {max_salary}', xy=(df.index[-1], max_salary), xytext=(df.index[-1], max_salary - 5000),
             arrowprops=dict(facecolor='green', shrink=0.05))

plt.show()
```
